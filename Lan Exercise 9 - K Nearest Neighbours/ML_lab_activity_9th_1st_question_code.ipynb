```python
import os
import numpy as np
from collections import Counter
```

Converting a 32×32 text image into a vector

```python
def img2vector(filename):
    vec=np.zeros((1,1024))
    with open(filename) as f:
        for i in range(32):
            line=f.readline()
            for j in range(32):
                vec[0,32*i+j]=int(line[j])
    return vec
```

**Each handwritten digit is stored as a 32×32 grid of 0s and 1s.**

Now we create the kNN Classifier

```python
def knn(test,data,labels,k=3):
    diff=data-test
    dist=np.sqrt((diff**2).sum(axis=1))
    idx=dist.argsort()
    vote=[]
    for i in range(k):
        vote.append(labels[idx[i]])
    return Counter(vote).most_common(1)[0][0]
```

It calculates the Euclidean distance between the test image and every training image, selects the nearest k neighbours, and predicts the digit based on majority voting.

NOW we are loadin the training dataset

```python
training_path="machinelearninginaction/Ch02/trainingDigits"
#Reads every handwritten digit from the training folder.
files=os.listdir(training_path)
m=len(files)
train=np.zeros((m,1024))
labels=[]
for i,file in enumerate(files):
    label=int(file.split("_")[0])
    labels.append(label)
    train[i]=img2vector(os.path.join(training_path,file))
```

Testing the classifier

```python
test_path="machinelearninginaction/Ch02/testDigits"
test_files=os.listdir(test_path)
errors=0
for file in test_files:
    actual=int(file.split("_")[0])
    vec=img2vector(os.path.join(test_path,file))
    pred=knn(vec,train,labels,3)
    print(f"{file}  Predicted:{pred}  Actual:{actual}")
    if pred!=actual:
        errors+=1
```

Then calculating the accuracy

Displays the classifiers performance by showing the total number of test images correct predictions incorrect predictions and the overall classification accuracy.

```python
total=len(test_files)

print("\nTotal Test Images :",total)
print("Incorrect :",errors)
print("Correct :",total-errors)
print("Accuracy :",((total-errors)/total)*100,"%")
```
