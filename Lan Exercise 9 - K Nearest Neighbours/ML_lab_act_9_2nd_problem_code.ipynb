```python
import pandas as pd
import numpy as np
```

Create the dataset from the tabular colummn

```python
data={
    "Brightness":[40,50,60,10,70,60,25],
    "Saturation":[20,50,90,25,70,10,80],
    "Class":["Red","Blue","Blue","Red","Blue","Red","Blue"]
}
df=pd.DataFrame(data)
df
```

# Define the new entry

Defines the new data point whose Euclidean distance from every existing record will be calculated.

```python
new_point=np.array([50,40])
print("New Entry")
print("Brightness =",new_point[0])
print("Saturation =",new_point[1])
```

Calculate Euclidean distance

i had a lot of difficulty learning this concept

```python
dist=[]
for i in range(len(df)):
    point=np.array([df.loc[i,"Brightness"],df.loc[i,"Saturation"]])
    d=np.sqrt(np.sum((point-new_point)**2))
    dist.append(d)
df["Euclidean Distance"]=dist
df
```

Sort by nearest neighbour

```python
df=df.sort_values("Euclidean Distance")
df
```
